```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ReLU(x):
    return np.maximum(x,0)
X=np.linspace(-10,10)
y_relu=ReLU(X)
plt.plot(y_relu)
```

```python
print("ReLU :", y_relu)
```

```python
def LeakyReLU(x):
    return max(x,x*0.1)
y = [LeakyReLU(i) for i in X]

plt.plot(y)
```

```python
print("ReLU :", y_relu)
print("LeakyReLU :", y)
```
